```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
```

```python
query = 'EU recession lang:en until:2022-11-11 since:2013-12-16 -filter:links -filter:replies'
tweets = []
limits = 200000
```

```python
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
  if len(tweets) == limits:
    break
  else:
    tweets.append([tweet.date, tweet.username, tweet.content])
```

```python
df1 = pd.DataFrame(tweets, columns =['Date', 'User', 'Tweet'])
#df1['Date'] = pd.to_datetime(df1['Date'], format='%Y-%m-%d')
#df1.to_csv('Tweet.csv')
pd.set_option('display.max_colwidth', None)

print(df1)
```

# Cleaning Tweets

```python
import re 
```

```python
def clean_tweet(tweet):
  if type (tweet) == float:
    return ""
  temp = tweet.lower()
  temp = re.sub("'", "", temp) #to avoid removing contractions in english
  temp = re.sub(r"(@[A-Za-z0-9_]+)|[^\w/\s]|#|http\S+","", temp)
  temp = re.sub('[()!?]', ' ', temp)
  temp = re.sub('\[.*?\]',' ', temp)
  temp = re.sub("[^a-z0-9]", " ", temp)
  return temp 
```

```python
clean_tweets = []
for tw in df1['Tweet']:
  results = clean_tweet(tw)
  clean_tweets.append(results)

df1['twee']= clean_tweets

#df1.to_csv("Cleaned_Tweet.csv")

df1
```

```python
df2= df1[['Date','User','twee']]
df2 = df2.sort_values(by='Date', ascending=True)
df2
```

```python
#df2['sentiment'] = np.nan
df2['twee'] = df2['twee'].replace('  ',' ')
df2['twee'] = df2['twee'].replace('   ',' ')
df2.head()
```

```python
from datetime import datetime
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
df2['Date'] = df2['Date'].dt.strftime('%Y-%m-%d')
```
